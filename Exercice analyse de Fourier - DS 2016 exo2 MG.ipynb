```python
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
from sympy import fourier_transform, Piecewise
from sympy.plotting import plot
from sympy.abc import t
from math import pi

#import sys
%config InlineBackend.figure_formats = {'png', 'retina'} # makes nicer graphics
%matplotlib inline
```

```python

t,n = symbols('t,n')
integrate(2/pi*2*t*sin(n*t), (t,-pi,pi))
```

```python

t = np.linspace(-np.pi, np.pi, 256, endpoint=True)
f0 = 0*t
f1 = 4*np.sin(t)
f2 = 4*(np.sin(t) - np.sin(2*t)/2)
f3 = 4*(np.sin(t) - np.sin(2*t)/2 + np.sin(3*t)/3)
f4 = 4*(np.sin(t) - np.sin(2*t)/2 + np.sin(3*t)/3 - np.sin(4*t)/4)
f5 = 4*(np.sin(t) - np.sin(2*t)/2 + np.sin(3*t)/3 - np.sin(4*t)/4 + np.sin(5*t)/5)
plt.figure(figsize=(20,10))
init_printing(use_latex='mathjax')
plt.plot(t,2*t,label="fonction originale") # fonction originale
plt.plot(t,f0,label="approximation ordre 0")
plt.plot(t,f1,label="approximation ordre 1")
plt.plot(t,f2,label="approximation ordre 2")
plt.plot(t,f3,label="approximation ordre 3")
plt.plot(t,f4,label="approximation ordre 4")
plt.plot(t,f5,label="approximation ordre 5")
plt.ylim([-8,13])
plt.legend(loc='upper left')
plt.show()
```
